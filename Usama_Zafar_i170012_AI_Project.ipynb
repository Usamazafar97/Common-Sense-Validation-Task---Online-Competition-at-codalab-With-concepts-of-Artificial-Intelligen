```
import math
from pytorch_pretrained_bert import OpenAIGPTTokenizer, OpenAIGPTModel, OpenAIGPTLMHeadModel
# Load pre-trained model (weights)
model_gpt = OpenAIGPTLMHeadModel.from_pretrained('openai-gpt')
# model_gpt.eval()
# Load pre-trained model tokenizer (vocabulary)
tokenizer_gpt = OpenAIGPTTokenizer.from_pretrained('openai-gpt')
```

```
def perplexity_score(sentence):
    tokenize_input = tokenizer_gpt.tokenize(sentence)
    tensor_input = torch.tensor([tokenizer_gpt.convert_tokens_to_ids(tokenize_input)])
    loss=model_gpt(tensor_input, lm_labels=tensor_input)
    
    return math.exp(int(loss))
```

```
import numpy as np

output_array = np.zeros(shape=(1000,2))
```

```
correct = 0
count = 0
for i in range(len(np_array)):
  
  sent1_score = perplexity_score(np_array[i][1])
  sent2_score = perplexity_score(np_array[i][2])

  # sent1_score = score(np_array[i][1])
  # sent2_score = score(np_array[i][2])

  # print("Sent1: ", sent1_score)
  # print("Sent2: ", sent2_score)
  
  if(sent1_score < sent2_score):
    print(i+1, ": ",1)
    if(np_array_answer[i][1] == 1):
      correct+=1
    output_array[i] = [np_array_answer[i][0],1]

  else:
    print(i+1, ": ",0)
    if(np_array_answer[i][1] == 0):
      correct+=1
    output_array[i] = [np_array_answer[i][0],0]
  print(output_array[i])
  count+=1

  print("Accuracy: " , correct/count)
```
